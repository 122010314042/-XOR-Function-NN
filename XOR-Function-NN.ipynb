```python
# The training data.
X = np.array([
[0, 0],
[0, 1],
[1, 0],
[1, 1]
])
# The labels for the training data.
y = np.array([
[0],
[1],
[1],
[0]
])
```

```python
W11 = np.asarray([1,1])
W12 = np.asarray([1,1])
W2 = np.asarray([-1,1])
B1 = np.asarray([-1.5,-0.5])
B2 = np.asarray([-0.5])
```

```python
def summation(x,Wt1,Wt2,Bs):
    sumh=[0,0]
    for i in range(len(x)):
        sumh[i]=(x[0]*Wt1[i]+x[1]*Wt2[i])+Bs[i]
    return sumh
```

```python
sumv=summation(x,W11,W12,B1)
```

```python
print(sumv)
```

```python
def Heaviside (net):
    if net >= 0:
        return 1
    else:
        return 0
```

```python
hidden=[0,0]
for i in range(len(sumv)):
    hidden[i]=Heaviside(sumv[i])
```

```python
print(hidden)
```

```python
sumo=np.dot(hidden,W2)+B2[0]
y_pred=Heaviside(sumo)
```

```python
print(y_pred)
```

```python
y_pred=[]
for i in range(X.shape[0]):
    x=X[i]
    sumv=summation(x,W11,W12,B1)
    hidden=[0,0]
    for i in range(len(sumv)):
        hidden[i]=Heaviside(sumv[i])
    sumo=np.dot(hidden,W2)+B2[0]
    y_pred.append(Heaviside(sumo))
```
